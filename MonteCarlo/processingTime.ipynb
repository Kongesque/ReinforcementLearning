# Processing Time Simulation

```python
import numpy as np
import matplotlib.pyplot as plt

simulaion = 1_000_000

machine1Time = np.random.uniform(1, 5, simulaion) # Machine 1 processes in 1 to 5 hours
machine2Time = np.random.uniform(2, 6, simulaion) # Machine 2 processes in 2 to 6 hours

processingTime = machine1Time + machine2Time
```

```python
plt.figure(figsize=(8, 4)) 
plt.hist(processingTime, bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(9, color='red', linestyle='--', linewidth=1.5, label="Deadline (9 hours)")
plt.title("Distribution of Total Processing Time")
plt.xlabel("Processing Time (hours)")
plt.ylabel("Density")
plt.legend()
plt.tight_layout()
plt.show()

probability_exceeding_deadline = np.mean(processingTime > 9) * 100
print(f"Percentage of total processing times exceeding 9 hours: {probability_exceeding_deadline:.2f}%")
```
